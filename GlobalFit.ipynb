# Global fit to semi-leptonic operators from $b\to s\mu^+\mu^-$ transitions

Example notebook to reproduce the numerics in the paper [arXiv:1703.09189v2](https://arxiv.org/abs/1703.09189v2) by W. Altmannshofer, C. Niehoff, P. Stangl, and D. M. Straub

## Imports and notebook settings

```python
import flavio.plots
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
from fit import fit
flavio.__version__
```

## Preparing the fit 

The following command will try to read the theory uncertainties & correlations from a data file. If it fails (depending on your system), the `make_measurement` command below will take much longer to execute.

```python
fit.load_sm_covariance('sm_covariance.p')
```

```python
%time fit.make_measurement(threads=4)
```

## Wilson coefficient plot

```python
%%time

# setting some options
plt.figure(figsize=(4,4))
opt = dict(x_min=-2, x_max=0.5, y_min=-1.25, y_max=1.25, n_sigma=(1, 2, 3), interpolation_factor=20)

flavio.plots.likelihood_contour(fit.log_likelihood, threads=4, **opt)

# labels, legend, etc.
flavio.plots.flavio_branding(y=0.07, x=0.05)
plt.axhline(0, c='k', lw=0.2)
plt.axvline(0, c='k', lw=0.2)

plt.xlabel(r'$\mathrm{Re}\,C_9^\mathrm{NP}$')
plt.ylabel(r'$\mathrm{Re}\,C_{10}^\mathrm{NP}$')
```
